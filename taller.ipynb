```python
# ==============================
# RETO 1 - RUTA ÓPTIMA
# ==============================
paquetes = []
TARIFA_BASE = 5000  # tarifa por kg en pesos colombianos
def ingresar_datos():
   while len(paquetes) < 5:
       print(f"\nRegistro #{len(paquetes)+1}")
       try:
           peso = float(input("Ingrese el peso del paquete (kg): "))
           if peso <= 0:
               print("Error: El peso no puede ser negativo o cero.")
               continue
           destino = input("Ingrese el destino del paquete: ")
           # Clasificación automática
           if peso < 1:
               tipo = "Documento"
           elif peso < 10:
               tipo = "Paquetería"
           else:
               tipo = "Carga"
           costo = peso * TARIFA_BASE
           paquete = {
               "peso": peso,
               "destino": destino,
               "tipo": tipo,
               "costo": costo
           }
           paquetes.append(paquete)
           print("Paquete registrado correctamente.")
       except:
           print("Error: Ingrese un valor numérico válido para el peso.")
def ver_reporte():
   if len(paquetes) < 5:
       print("Debe ingresar al menos 5 paquetes antes de ver el reporte.")
       return
   # Ordenar por peso de mayor a menor
   paquetes.sort(key=lambda x: x["peso"], reverse=True)
   print("\n===== MANIFIESTO DE CARGA =====")
   total_envio = 0
   for paquete in paquetes:
       print(f"Destino: {paquete['destino']} | "
             f"Peso: {paquete['peso']} kg | "
             f"Tipo: {paquete['tipo']} | "
             f"Costo: ${paquete['costo']}")
       total_envio += paquete["costo"]
   print(f"\nTotal general de envíos: ${total_envio}")
# ==============================
# MENÚ INTERACTIVO
# ==============================
while True:
   print("\n===== MENÚ RUTA-ÓPTIMA =====")
   print("1. Ingresar paquetes")
   print("2. Ver manifiesto de carga")
   print("3. Salir")
   opcion = input("Seleccione una opción: ")
   if opcion == "1":
       ingresar_datos()
   elif opcion == "2":
       ver_reporte()
   elif opcion == "3":
       print("Saliendo del sistema logístico...")
       break
   else:
       print("Opción inválida. Intente nuevamente.")
```

```python
# ===================================
# RETO 4 - FIDELIZA-PUNTOS
# ===================================
clientes = []
def registrar_clientes():
   while len(clientes) < 5:
       print(f"\nRegistro cliente #{len(clientes)+1}")
       nombre = input("Ingrese el nombre del cliente: ").strip()
       try:
           compra = float(input("Ingrese el valor total de compra: "))
           if compra < 0:
               print("Error: El valor de compra no puede ser negativo.")
               continue
           # Asignación de categoría
           if compra < 100000:
               categoria = "Bronce"
               bono = 0
           elif compra < 500000:
               categoria = "Plata"
               bono = 0
           else:
               categoria = "Oro"
               bono = compra * 0.10  # 10% de bono
           cliente = {
               "nombre": nombre,
               "compra": compra,
               "categoria": categoria,
               "bono": bono
           }
           clientes.append(cliente)
           print("Cliente registrado correctamente.")
       except:
           print("Error: Ingrese un valor numérico válido para la compra.")
def ver_reporte():
   if len(clientes) < 5:
       print("Debe registrar al menos 5 clientes antes de ver el reporte.")
       return
   # Ordenar alfabéticamente por nombre
   clientes.sort(key=lambda x: x["nombre"].lower())
   print("\n===== REPORTE DE CLIENTES - FIDELIZA PUNTOS =====")
   total_ventas = 0
   total_bonos = 0
   for cliente in clientes:
       print(f"Nombre: {cliente['nombre']} | "
             f"Compra: ${cliente['compra']:.0f} | "
             f"Categoría: {cliente['categoria']} | "
             f"Bono: ${cliente['bono']:.0f}")
       total_ventas += cliente["compra"]
       total_bonos += cliente["bono"]
   print("\n----- RESUMEN GENERAL -----")
   print(f"Total ventas: ${total_ventas:.0f}")
   print(f"Total bonos entregados: ${total_bonos:.0f}")
# ===================================
# MENÚ INTERACTIVO
# ===================================
while True:
   print("\n===== MENÚ FIDELIZA-PUNTOS =====")
   print("1. Registrar clientes")
   print("2. Ver reporte")
   print("3. Salir")
   opcion = input("Seleccione una opción: ")
   if opcion == "1":
       registrar_clientes()
   elif opcion == "2":
       ver_reporte()
   elif opcion == "3":
       print("Saliendo del sistema de fidelización...")
       break
   else:
       print("Opción inválida. Intente nuevamente.")
```
